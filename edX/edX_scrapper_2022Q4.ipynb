```python
csv_file = open('courses_whole.csv', 'w', encoding = 'utf-8' )

writer = csv.writer(csv_file)

writer.writerow([
    'course_link', 
    'title', 'short_description', 'length', 'detail', 
    'institution', 'level', 'prerequisite', 'language', 'transcript', 
    'instructor', 'verified_track', 'audit_track'])
```

```python
max_page = 42

page_number = list(range(11, max_page+1))

page_list = []

web_page = 'https://edx.org/search?tab=course&page='
page_list = [web_page + str(i) for i in page_number]
```

```python
course_url=[]

for url in page_list:
    driver = webdriver.Chrome("/Users/Nur Adhyaksa Hamid/Dropbox/My PC (LAPTOP-PF5V2EGV)/Downloads/chromedriver_win32_107/chromedriver")
    driver.get(url) 
    sleep(randint(5,15))
    WebDriverWait(driver, 30).until(EC.visibility_of_element_located((By.XPATH, '/html/body/div[1]/div[1]/div/main/div/div[4]/div/div/div/div[2]/div/div[1]/div/a')))
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    hrefs = [item.get("href") for item in soup.find_all("a")]
    
    hrefs_final = list(dict.fromkeys(hrefs))
    hrefs_final = list(filter(None, hrefs_final)) 

    href_final = [x for x in hrefs_final if x.startswith('/course/')]
    
    main_page = 'https://edx.org'
    course_list=[main_page + s for s in href_final]
    
    time.sleep(10)

    course_url.append(course_list)
    
course_url.append(course_list)

#data = {'URLs': course_list}

#url_list = pd.DataFrame.from_dict(data)

#url_list

#df.to_csv('output.csv', index = 0)
```

```python
# get title of the course
def get_title():
    try:
        title = driver.find_element_by_xpath('.//div[contains(@class, "col-md-7 pr-4")]//h1').text
    except:
        title = 'Missing'
    finally:
        return title

# get course's short description
def get_short_description():
    try:
        short_description = driver.find_element_by_xpath('.//div[contains(@class, "col-md-7 pr-4")]//p').text
    except:
        short_description = 'Missing'
    finally:
        return short_description

# get course completion length
def get_length():
    try:
        length = driver.find_element_by_xpath('.//div[contains(@class, "ml-3")]//div[contains(@class, "h4 mb-0")]').text
    except:
        length = 'Missing'   
    finally:
        return length

# get dedication hour
def get_detail():
    try:
        detail = driver.find_element_by_xpath('.//div[contains(@class, "ml-3")]//div[contains(@class, "small")]').text
    except:
        detail = 'Missing'
    finally:
        return detail
    
# get providing institution
def get_institution():
    try:
        institution = driver.find_element_by_xpath('.//div[contains(@class, "at-a-glance")]//a[contains(@class, "muted-link")]').text
    except:
        institution = 'Missing'
    finally:
        return institution
    
# get course's level
def get_level():
    try:
        level = driver.find_element_by_xpath('.//div[contains(@class, "at-a-glance")]//ul[@class="mb-0 pl-3 ml-1"]/li[3]').text
    except:
        level = 'Missing'
    finally:
        return level
    
# get course prerequisite
def get_prerequisite():
    try:
        prerequisite = driver.find_element_by_xpath('.//div[contains(@class, "at-a-glance")]//ul[@class="mb-0 pl-3 ml-1"]/li[4]').text
    except:
        prerequisite = 'Missing'
    finally:
        return prerequisite
    
# get course language
def get_language():
    try:
        language = driver.find_element_by_xpath('.//div[contains(@class, "at-a-glance")]//ul[@class="pl-3 ml-1 mb-0"]/li[1]').text
    except:
        language = 'Missing'
    finally:
        return language
    
# get course transcript
def get_transcript():
    try:
        transcript = driver.find_element_by_xpath('.//div[contains(@class, "at-a-glance")]//ul[@class="pl-3 ml-1 mb-0"]/li[2]').text
    except:
        transcript = 'Missing'
    finally:
        return transcript

# get course instructor
def get_instructor():
    try:
        instructor = driver.find_element_by_xpath('.//div[contains(@class, "instructor-cards d-flex flex-wrap justify-content-center mx-auto")]').text
    except:
        instructor = 'Missing'
    finally:
        return instructor
    
# get verified track cost
def get_verified():
    try:
        verified_track = driver.find_element_by_xpath('.//div[contains(@class, "track-comparison-table-wrapper")]/table/tbody/tr[2]/td[1]/p').text
    except:
        verified_track = 'Missing'
    finally:
        return verified_track
    
# get audit track cost
def get_audit():
    try:
        audit_track = driver.find_element_by_xpath('.//div[contains(@class, "track-comparison-table-wrapper")]/table/tbody/tr[2]/td[2]/p').text
    except:
        audit_track = 'Missing'
    finally:
        return audit_track
```

```python
for course_link in course_url:
    course_dict = {}
    driver = webdriver.Chrome('/Users/Nur Adhyaksa Hamid/Dropbox/My PC (LAPTOP-PF5V2EGV)/Downloads/chromedriver_win32_107/chromedriver')
    driver.get(course_link)
    
        
    course_dict['link'] = course_link
    course_dict['title'] = get_title()
    course_dict['short_description'] = get_short_description()
    course_dict['length'] = get_length()
    course_dict['detail'] = get_detail()
    course_dict['institution'] = get_institution()
    course_dict['level'] = get_level()
    course_dict['prerequisite'] = get_prerequisite()
    course_dict['language'] = get_language()
    course_dict['transcript'] = get_transcript()
    course_dict['instructor'] = get_instructor()
    course_dict['verified_track'] = get_verified()
    course_dict['audit_track'] = get_audit()
    
    writer.writerow(course_dict.values())
    driver.close()

#close the csv once all course info is scraped
csv_file.close()
```
