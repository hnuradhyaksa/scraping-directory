```python
import numpy as np
import pandas as pd

data = []

header = {'User-Agent': 'python-requests/2.26.0', 
          'Accept-Encoding': 'gzip, deflate, br', 
          'Accept': '*/*', 'Connection': 'keep-alive'
         }
```

```python
source = pd.read_csv(
    r'C:\Users\Nur Adhyaksa Hamid\OneDrive\Documents\Project\MOOC\df_complete.csv'
)

course_link = source['Link']

#course_link = course_link.head(5)

web_page = 'https://www.classcentral.com'
course_link = [web_page + str(i) for i in course_link]
```

```python
course_link
```

```python
for current_page in course_link:
    page = requests.get(
        current_page, 
        headers = header)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    try :
        course_title = soup.find_all('h1', class_="head-2 medium-up-head-1 small-down-margin-bottom-xsmall")[0].get_text()
    except IndexError:
        course_title = 'Not Found'
        
    try :
        course_cost = soup.find_all('span', class_="text-2 line-tight")[1].get_text()
        
    except IndexError:
        course_cost = 'Not Found'
                      
    data.append({
        'Title' : course_title,
        'Cost' : course_cost
    })

df = pd.DataFrame(data)
```
